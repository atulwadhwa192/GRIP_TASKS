```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
dist_from_centr=[]
k=list(range(1,10))
for i in k:
    k_model=KMeans(n_clusters=i)
    k_model.fit(df)
    dist_from_centr.append(k_model.inertia_)
```

```python
dist_from_centr
```

```python
plt.figure(figsize=(10,7))
sns.set_context(context='poster')
plt.plot(k,dist_from_centr)
plt.xlabel('Clusters value')
plt.ylabel('Distance from centroid')
plt.legend(('Distance',),loc='best')
plt.show()
```

```python
plt.plot(k,dist_from_centr)
plt.xlabel('Clusters value')
plt.ylabel('Distance from centroid')
plt.plot([k[0],k[-1]],[dist_from_centr[0],dist_from_centr[-1]])
plt.legend(('Distance','Longest Dist'),loc='best')
```

```python
model=KMeans(n_clusters=3,random_state=42)
```

```python
model.fit(df)
y_pred=model.predict(df)
#model.fit_predict(df)
#model.labels_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(df.iloc[y_pred == 0, 0], df.iloc[y_pred == 0, 1], 
            s = 50, c = 'red', label = 'Iris-setosa')
plt.scatter(df.iloc[y_pred == 1, 0], df.iloc[y_pred == 1, 1], 
            s = 50, c = 'blue', label = 'Iris-versicolour')
plt.scatter(df.iloc[y_pred == 2, 0], df.iloc[y_pred == 2, 1],
            s = 50, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:,1], 
            s = 50, c = 'yellow', label = 'Centroids')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
```
